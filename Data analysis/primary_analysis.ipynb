```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
mental_health = pd.read_csv("Teen_Mental_Health_Dataset.csv")
mental_health.head(5)

mental_health.info()
# ? No nulls in the base
mental_health.describe()
# ? General view of the base
mental_health.shape
# ? 1200 x 13
mental_health.dtypes
# ? Everything with the data types = OK
```

```python
# Demographic of ages of the users in the data base
print(mental_health["age"].unique())
age_counter = Counter(mental_health["age"])
xs = range(13, 19)
ys = [age_counter[x] for x in xs]
color = "#BEEDAA"
plt.bar(xs, ys, color=color, edgecolor="black")
plt.axis([12, 19, 0, 210])
plt.title("Demographic of ages in the dataset")
plt.show()
```

```python
# Comparing genders

gender_count = Counter(mental_health["gender"])
print(gender_count)

labels = list(gender_count.keys())
values = list(gender_count.values())
colors = ["#0C1B33", "#F05365"]
plt.title("Comparassion of genders")
plt.pie(values, labels = labels, startangle = 90, colors=colors, wedgeprops={'edgecolor': 'black', 'linewidth': 0.8})
plt.show()
```

```python
# Analysis of discret numbers

plt.hist(mental_health["daily_social_media_hours"], color="#F05365", edgecolor="black", bins=7)
plt.title("Daily social media hours")
plt.show()

plt.hist(mental_health["sleep_hours"], color="#8B80F9", edgecolor="black", bins=10)
plt.title("Sleep hours")
plt.show()

plt.hist(mental_health["screen_time_before_sleep"], color="#0C1B33", edgecolor="black", bins=5)
plt.title("Screen time before sleep")
plt.show()

plt.hist(mental_health["academic_performance"], color="#275DAD", edgecolor="black", bins=8)
plt.title("Academic performance")
plt.show()

plt.hist(mental_health["physical_activity"], color="#5F0F40", edgecolor="black", bins=8)
plt.title("Physical activity")
plt.show()
```

```python
# Comparing platform usage

platform_usage_count = Counter(mental_health["platform_usage"])

labels = list(platform_usage_count.keys())
values = list(platform_usage_count.values())
colors = ["#E4FF1A", "#275DAD", "#8B80F9"]
explode = [0.02, 0.02, 0.02]
plt.title("Comparassion of platform usage")
plt.pie(values, labels = labels, startangle = 90, colors=colors, explode=explode, wedgeprops={'edgecolor': 'black', 'linewidth': 0.8})
plt.show()
```

```python
#social_interaction_level

social_interaction_level_counter = Counter(mental_health["social_interaction_level"])
print(social_interaction_level_counter)

labels = list(social_interaction_level_counter.keys())
values = list(social_interaction_level_counter.values())

colors = ["#F05365", "#8B80F9", "#92B4A7"]
explode = [0.02, 0.02, 0.02]

plt.title("Comparassion of social interaction levels")
plt.pie(values, labels = labels, startangle = 90, colors=colors, explode=explode, wedgeprops={'edgecolor': 'black', 'linewidth': 0.8})
plt.show()
```

```python
# Analysis of stress_level, anxiety_level and addiction_level

stress_level_counter = Counter(mental_health["stress_level"])
xs = range(0, 11)
ys = [stress_level_counter[x] for x in xs]
plt.bar(xs, ys, color="#9999C3", edgecolor="black")
plt.axis([0, 11, 0, 150])
plt.title("Stress levels")
plt.show()

anxiety_level_counter = Counter(mental_health["anxiety_level"])
xs = range(0, 11)
ys = [anxiety_level_counter[x] for x in xs]
plt.bar(xs, ys, color="#E88D67", edgecolor="black")
plt.axis([0, 11, 0, 150])
plt.title("Anxiety levels")
plt.show()

addiction_level_counter = Counter(mental_health["addiction_level"])
xs = range(0, 11)
ys = [addiction_level_counter[x] for x in xs]
plt.bar(xs, ys, color="#C0E6DE", edgecolor="black")
plt.axis([0, 11, 0, 150])
plt.title("Addiction levels")
plt.show()
```

```python
# depression label

depression_label_counter = Counter(mental_health["depression_label"])
print(depression_label_counter)

labels = list(depression_label_counter.keys())
values = list(depression_label_counter.values())

colors = ["#F05365", "#8B80F9"]
explode = [0.02, 0.02]

plt.title("Porcentange of depressed students")
plt.pie(values, labels = labels, startangle = 90, colors=colors, explode=explode, wedgeprops={'edgecolor': 'black', 'linewidth': 0.8})
plt.show()
```

```python
anxiety_corr = mental_health["daily_social_media_hours"].corr(mental_health["anxiety_level"])
print(anxiety_corr)
```

```python
# !
mental_health["usage_group"] = pd.cut(
    mental_health["daily_social_media_hours"],
    bins=[0, 2, 4, 6, 24],
    labels=["low", "moderate", "high", "very_high"]
)

print(mental_health.groupby("usage_group", observed=True)["anxiety_level"].mean())


mental_health["sleep_group"] = pd.cut(
    mental_health["sleep_hours"],
    bins=[0, 4, 6, 10],
    labels=["very low", "low", "normal"]
)

print(mental_health.groupby("sleep_group", observed=True)["anxiety_level"].mean())


mental_health["screen_time_before_sleep_group"] = pd.cut(
    mental_health["screen_time_before_sleep"],
    bins=[0, 1, 4],
    labels=["low", "high"],
    include_lowest=True
)

print(mental_health.groupby("screen_time_before_sleep_group", observed=True)["anxiety_level"].mean())

```
