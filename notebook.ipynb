```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
mental_health = pd.read_csv("Teen_Mental_Health_Dataset.csv")
mental_health.head(5)
```

```python
mental_health.shape
```

```python
mental_health["age"].unique()
```

```python
# Demographic of ages of the users in the data base

age_counter = Counter(mental_health["age"])
xs = range(13, 19)
ys = [age_counter[x] for x in xs]
plt.bar(xs, ys)
plt.axis([12, 19, 0, 210])
plt.title("Demographic of ages in the dataset")
plt.show()
```

```python
# Comparing genders

gender_count = Counter(mental_health["gender"])
print(gender_count)

labels = list(gender_count.keys())
values = list(gender_count.values())
colors = ["#0C1B33", "#F05365"]
plt.title("Comparassion of genders")
plt.pie(values, labels = labels, startangle = 90, colors=colors)
plt.show()
```

```python
# Histogram of daily social media hours

plt.hist(mental_health["daily_social_media_hours"], color="#F05365", edgecolor="black", bins=10)
plt.title("Daily social media hours")
plt.show()
plt.hist(mental_health["sleep_hours"], color="#8B80F9", edgecolor="black", bins=10)
plt.title("Sleep hours")
plt.show()
plt.hist(mental_health["screen_time_before_sleep"], color="#0C1B33", edgecolor="black", bins=16)
plt.title("Screen time before sleep")
plt.show()
plt.hist(mental_health["academic_performance"], color="#275DAD", edgecolor="black", bins=16)
plt.title("Academic performance")
plt.show()
plt.hist(mental_health["physical_activity"], color="#5F0F40", edgecolor="black", bins=16)
plt.title("Physical activity")
plt.show()
```

```python
# Comparing platform usage

platform_usage_count = Counter(mental_health["platform_usage"])

labels = list(platform_usage_count.keys())
values = list(platform_usage_count.values())
colors = ["#E4FF1A", "#275DAD", "#8B80F9"]
explode = [0.02, 0.02, 0.02]
plt.title("Comparassion of platform usage")
plt.pie(values, labels = labels, startangle = 90, colors=colors, explode=explode)
plt.show()
```

```python
#social_interaction_level

social_interaction_level_counter = Counter(mental_health["social_interaction_level"])
print(social_interaction_level_counter)

labels = list(social_interaction_level_counter.keys())
values = list(social_interaction_level_counter.values())

colors = ["#F05365", "#8B80F9", "#92B4A7"]
explode = [0.02, 0.02, 0.02]

plt.title("Comparassion of platform usage")
plt.pie(values, labels = labels, startangle = 90, colors=colors, explode=explode)
plt.show()
```

```python
# stress_level	anxiety_level	addiction_level

stress_level_counter = Counter(mental_health["stress_level"])
xs = range(0, 11)
ys = [stress_level_counter[x] for x in xs]
plt.bar(xs, ys)
plt.axis([0, 11, 0, 150])
plt.title("Stress levels")
plt.show()

anxiety_level_counter = Counter(mental_health["anxiety_level"])
xs = range(0, 11)
ys = [anxiety_level_counter[x] for x in xs]
plt.bar(xs, ys, color="red")
plt.axis([0, 11, 0, 150])
plt.title("Anxiety levels")
plt.show()

addiction_level_counter = Counter(mental_health["addiction_level"])
xs = range(0, 11)
ys = [addiction_level_counter[x] for x in xs]
plt.bar(xs, ys, color="green")
plt.axis([0, 11, 0, 150])
plt.title("Addiction levels")
plt.show()
```

```python
# depression label


depression_label_counter = Counter(mental_health["depression_label"])
print(depression_label_counter)

labels = list(depression_label_counter.keys())
values = list(depression_label_counter.values())

colors = ["#F05365", "#8B80F9"]
explode = [0.02, 0.02]

plt.title("Comparassion of platform usage")
plt.pie(values, labels = labels, startangle = 90, colors=colors, explode=explode)
plt.show()
```
